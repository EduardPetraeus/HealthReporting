Build a core data model for the food consumed

```python
%sql
CREATE OR REPLACE TABLE workspace.default.DailyMeal_Clean AS
WITH deduped_food AS (
  SELECT
    *,
    ROW_NUMBER() OVER (
      PARTITION BY
        date,
        meal_type,
        title
      ORDER BY date DESC
    ) AS rn
  FROM workspace.default.food
)
SELECT
  date,
  meal_type,
  title,
  brand,
  serving_name,
  amount,
  amount_in_grams,
  calories,
  carbs,
  carbs_fiber,
  carbs_sugar,
  cholesterol,
  fat,
  fat_saturated,
  fat_unsaturated,
  potassium,
  protein,
  sodium,
  current_timestamp() AS load_datetime
FROM deduped_food
WHERE rn = 1 AND date IS NOT NULL
```
